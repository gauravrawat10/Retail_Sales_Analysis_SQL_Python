focuses on answering business-driven questions using the Superstore
dataset. The analysis aims to identify loss-making areas, profitable segments,
and key factors affecting profitability.

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(
    r"C:\Users\2005r\OneDrive\Desktop\superstore_dataset_project\excel_cleaned_data_csv\Superstore_Cleaned_Final.csv"
)
```

```python
#Loss making orders 
loss_orders = df[df['Profit'] < 0]

loss_orders.shape
```

```python
#Products causing most losses
loss_products = (
    df.groupby('Sub-Category')['Profit']
      .sum()
      .sort_values()
)

loss_products.head(10)
```

```python
#DISCOUNT VS PROFIT IMPACT
discount_profit = (
    df.groupby('Discount')
      .agg(
          Orders=('Order ID', 'count'),
          Total_Profit=('Profit', 'sum'),
          Avg_Profit=('Profit', 'mean')
      )
)

discount_profit
```

```python
#MOST PROFITABLE CUSTOMERS
top_customers = (
    df.groupby('Customer Name')['Profit']
      .sum()
      .sort_values(ascending=False)
      .head(10)
)

top_customers.map('${:,.2f}'.format)
```

```python
shipping_profit = (
    df.groupby('Ship Mode')
      .agg(
          Orders=('Order ID', 'count'),
          Total_Profit=('Profit', 'sum')
      )
      .sort_values('Total_Profit', ascending=False)
)

shipping_profit['Total_Profit'] = shipping_profit['Total_Profit'].round(2)

shipping_profit

```
